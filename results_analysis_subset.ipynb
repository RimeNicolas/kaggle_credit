```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

from collections import Counter
```

```python
df_rand_forest = pd.read_csv('./csv_dir/credit_Rand_Forest_23_05_2020_15_14.csv')
df_adaboost = pd.read_csv('./csv_dir/credit_AdaBoost_23_05_2020_15_15.csv')
df_gradientboost = pd.read_csv('./csv_dir/credit_GradientBoost_23_05_2020_15_16.csv')
```

```python
df = pd.concat([df_rand_forest, df_adaboost, df_gradientboost], ignore_index=True)
def df_move_column(idx_i, idx_f, df):
    cols = list(df.columns.values)
    lr = cols.pop(idx_i)
    cols.insert(idx_f,lr)
    return df[cols]
df = df_move_column(-1,3,df)
df
```

```python
df_largest_AUC = df.nlargest(15,'AUC')
df_largest_AUC
```

```python
df_smallest_FN = df.nsmallest(15,'FN')
df_smallest_FN
```

```python
df_largest_acc = df.nlargest(15, 'accuracy')
df_largest_acc
```

```python
pd.merge(df_largest_AUC, df_smallest_FN, how='inner')
```
