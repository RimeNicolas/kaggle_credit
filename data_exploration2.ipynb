```python
from importlib import reload
import test_models as mod1
mod1 = reload(mod1)

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

from collections import Counter
```

```python
kwargs = mod1.yaml_loader('./config.yaml')
X, y = mod1.prepare_data(kwargs['dir_data'], 'creditcard.csv')
```

```python
idx_is_fraud  = [i for i in range(len(y)) if y[i] == 1]
idx_not_fraud = [i for i in range(len(y)) if y[i] == 0]
```

```python
def hist_data(X):
    X_len = len(X.T)
    print('number of figures : ', X_len)
    for i, value in enumerate(X.T):
        if i < X_len and i%3 == 0:
            plt.figure(i, figsize=(21,3))
            plt.subplot(131)
            plt.hist(value, bins = 50, density = True)
        if i < X_len and i%3 == 1:
            plt.subplot(132)
            plt.hist(value, bins = 50, density = True)
        if i < X_len and i%3 == 2:
            plt.subplot(133)
            plt.hist(value, bins = 50, density = True)
    
hist_data(X[idx_is_fraud])
X_not_fraud = X[idx_not_fraud]
hist_data(X_not_fraud[:len(idx_is_fraud)])
plt.show()
```

```python
plt.figure(100)
plt.hist(X[idx_not_fraud, 16], bins = 50, density=True, color='b', label = 'not fraud')
plt.hist(X[idx_is_fraud, 16], bins = 50, density=True, color='r', label = 'fraud')
plt.show()
```
