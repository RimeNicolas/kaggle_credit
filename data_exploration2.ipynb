```python
from importlib import reload
import test_models as mod1
mod1 = reload(mod1)

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

from collections import Counter
```

```python
kwargs = mod1.yaml_loader('./config.yaml')
X, y = mod1.prepare_data(kwargs['dir_data'], 'creditcard.csv')
```

```python
idx_is_fraud  = [i for i in range(len(y)) if y[i] == 1]
idx_not_fraud = [i for i in range(len(y)) if y[i] == 0]
```

```python
def hist_data(X1, X2):
    X_len = len(X1.T)
    print('number of figures : ', X_len)
    for i in range(len(X1.T)):
        if i < X_len and i%3 == 0:
            figure = plt.figure(i, figsize=(10,3))
            ax1 = plt.Axes(figure, [.1, .1, .25, .80])
            figure.add_axes(ax1)
            ax1.hist([X1[:,i], X2[:,i]], bins = 50, density = True, label=['is fraud', 'not fraud'])
            if i == 0:
                ax1.legend(loc="upper left")
        if i < X_len and i%3 == 1:
            ax2 = plt.Axes(figure, [.4, .1, .25, .80])
            figure.add_axes(ax2)
            ax2.hist([X1[:,i], X2[:,i]], bins = 50, density = True)
        if i < X_len and i%3 == 2:
            ax3= plt.Axes(figure, [.7, .1, .25, .80])
            figure.add_axes(ax3)
            ax3.hist([X1[:,i], X2[:,i]], bins = 50, density = True)
    plt.show()
    

X_not_fraud = X[idx_not_fraud]
X1 = X[idx_is_fraud]
X2 = X_not_fraud[:len(idx_is_fraud)]
hist_data(X1, X2)
```
