```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

from collections import Counter
```

```python
df_rand_forest1 = pd.read_csv('./csv_dir/credit_Rand_Forest_1.csv')
df_rand_forest1.insert(0, 'model', ['Rand_Forest']*len(df_rand_forest1['TP']), allow_duplicates=True)
df_rand_forest2 = pd.read_csv('./csv_dir/credit_Rand_Forest_22_05_2020_15_47.csv')
df_rand_forest3 = pd.read_csv('./csv_dir/credit_Rand_Forest_22_05_2020_19_35.csv')
df_adaboost1 = pd.read_csv('./csv_dir/credit_AdaBoost_22_05_2020_17_01.csv')
df_adaboost2 = pd.read_csv('./csv_dir/credit_AdaBoost_22_05_2020_21_39.csv')
```

```python
df_rand_forest = pd.concat([df_rand_forest1, df_rand_forest2, df_rand_forest3], ignore_index=True).sort_values(by=['n_estimators', 'max_depth'])
df_adaboost = pd.concat([df_adaboost1, df_adaboost2], ignore_index=True).sort_values(by=['n_estimators', 'learning_rate'])
df = pd.concat([df_rand_forest, df_adaboost], ignore_index=True)
def df_move_column(idx_i, idx_f, df):
    cols = list(df.columns.values)
    lr = cols.pop(idx_i)
    cols.insert(idx_f,lr)
    return df[cols]
df = df_move_column(-1,3,df)
df
```

```python
df_largest_AUC = df.nlargest(5,'AUC')
df_largest_AUC
```

```python
df_smallest_FN = df.nsmallest(5,'FN')
df_smallest_FN
```

```python
df_largest_acc = df.nlargest(5, 'accuracy')
df_largest_acc
```

```python
pd.merge(df_largest_AUC, df_smallest_FN, how='inner')
```
